Amazon has undergone significant evolution since its inception in 1994. Here's a brief summary of its journey through the years:

The company also launched Amazon Web Services (AWS) in 2006, a cloud computing platform that became a significant driver of revenue and profit.

4. **Acquisitions and Innovations (2010s):** Amazon made several notable acquisitions during this decade, including Zappos (2009), Twitch (2014), and Whole Foods Market (2017). It also continued to innovate with products like the Kindle e-reader and Echo smart speaker powered by the Alexa virtual assistant.

5. **E-Commerce Dominance (2010s):** Amazon solidified its position as a dominant force in e-commerce, capturing a large share of online retail sales in various markets globally. The company also faced criticism over its treatment of workers and antitrust concerns related to its market power.

6. **COVID-19 Pandemic (2020):** The COVID-19 pandemic led to a surge in demand for online shopping, benefiting Amazon significantly. The company faced challenges in meeting increased demand while implementing safety measures for its workforce.

7. **Continued Growth and Expansion (2021-2022):** Amazon continued to expand its business ventures, including investments in logistics, healthcare, and entertainment. It faced increased regulatory scrutiny and criticism over issues such as labor practices, privacy concerns, and environmental impact.

Throughout its history, Amazon has demonstrated a relentless focus on customer convenience, innovation, and long-term growth, making it one of the world's most valuable and influential companies.

```python
# Convert 'IMDb Rating' to numeric
df_prime['IMDb Rating'] = pd.to_numeric(df_prime['IMDb Rating'], errors='coerce')

# Group by 'Language' and calculate mean IMDb rating for the top 10
Lang_dis = df_prime.groupby('Language')['IMDb Rating'].mean().sort_values(ascending=False)[:10]

# Compute the number of bars
num_bars = len(Lang_dis)

# Set the angles for each bar
angles = np.linspace(0, 2 * np.pi, num_bars, endpoint=False).tolist()

# The width of each bar
width = np.pi / num_bars

# Initialize the figure with one subplot
fig, ax = plt.subplots(1, 1, figsize=(12, 8), subplot_kw=dict(polar=True))


# Remove the axis
ax.axis('off')

# Plot each bar on the polar plot
for angle, value, lang_name in zip(angles, Lang_dis, Lang_dis.index):
    # IMDb rating labels
    imdb_rating_label = f'{value:.1f}'
    if value > 6.4:
        bar_color = '#08AAE3'  # Change color if rating is above 6.4
    else:
        bar_color = '#233D53'
    ax.bar(angle, value, width=width, color=bar_color)
    # Language name labels
    ax.text(angle, 5, lang_name, ha='center', va='bottom', fontsize=8, fontfamily='serif', color='white', fontweight='bold')
    # IMDb rating labels above the axis
    ax.text(angle, value + 0.5, imdb_rating_label, ha='center', va='bottom', fontsize=9, fontfamily='serif', color='black')

# Set the label position
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(angles)
ax.set_xticklabels(Lang_dis.index, fontsize=10, fontfamily='serif')

# Add text for the insight
fig.text(1.1, 1.01, 'Insight', fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(1.1, 0.67, '''
Based on IMDb ratings, the highlighted movies
reflect the highest-rated films within
their respective languages. Interestingly,
the analysis reveals a notable
trend: regional language films in India tend
to receive higher ratings compared to movies in
commonly spoken language (Hindi).
This suggests a strong preference and appreciation
for regional cinema among audiences,
emphasizing the diverse cultural landscape and
the rich storytelling traditions present in Indian cinema..
'''
, fontsize=12, fontweight='light', fontfamily='serif')

import matplotlib.lines as lines

l1 = lines.Line2D([1, 1], [0, 1], transform=fig.transFigure, figure=fig, color='black', lw=0.7)
fig.lines.extend([l1])

ax.set_xticks(angles)
ax.set_xticklabels(Lang_dis.index, fontsize=10, fontfamily='serif')

plt.title('Language Distribution Based on IMDb Ratings', fontsize=12, fontweight='bold', fontfamily='serif', loc='center')

ax.tick_params(axis=u'both', which=u'both', length=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame for movies released between 1961 and 2000
df_1961_2000 = df_prime[df_prime['Year of Release'].between(1961, 2000)]

# Filter the DataFrame for movies released between 2000 and 2021
df_2000_2021 = df_prime[df_prime['Year of Release'].between(2000, 2021)]

# Count the number of releases for each time period
count_1961_2000 = len(df_1961_2000)
count_2000_2021 = len(df_2000_2021)

# Calculate the ratio of releases between the two time periods
ratio_1961_2000 = count_1961_2000 / (count_1961_2000 + count_2000_2021)
ratio_2000_2021 = count_2000_2021 / (count_1961_2000 + count_2000_2021)

# Group by 'Year of Release' and count the number of releases for each time period
count_by_year_1961_2000 = df_1961_2000.groupby('Year of Release').size().cumsum()
count_by_year_2000_2021 = df_2000_2021.groupby('Year of Release').size().cumsum()

# Plotting the area chart
fig, ax = plt.subplots(figsize=(12, 6))

ax.fill_between(count_by_year_1961_2000.index, count_by_year_1961_2000, color='#08AAE3', alpha=0.5, label='1961-2000')
ax.fill_between(count_by_year_2000_2021.index, count_by_year_2000_2021, color='#233D53', alpha=0.5, label='2000-2021')

# Adding labels and title
ax.set_title('Cumulative Distribution of Movie Releases between 1961-2000 and 2000-2021', fontsize=14, fontweight='bold', fontfamily='serif')
ax.set_xlabel('Year of Release', fontsize=12, fontfamily='serif')

# Remove grid lines and spines
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Add additional text
fig.text(0.13, 0.85, 'Movies added over time [Cumulative Total]', fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.13, 0.58,
'''Prime's peak global content was between 2000 and 2021,
indicating a significant emphasis on expanding its library
during this period. Particularly noteworthy is the
escalated focus on producing movie content in India
since the 2000s. The growth in movie offerings has
been notably more pronounced compared to the 90s.
''', fontsize=12, fontweight='light', fontfamily='serif')

fig.text(0.3, 0.2, "Movies - 90's", fontweight="bold", fontfamily='serif', fontsize=15, color='#233D53')
fig.text(0.7, 0.2, "Movies - 2000's", fontweight="bold", fontfamily='serif', fontsize=15, color='#221f1f')

ax.tick_params(axis=u'both', which=u'both', length=0)

# Move x-axis ticks to the right
ax.yaxis.tick_right()

# Display the plot
plt.tight_layout()
plt.show()
```
